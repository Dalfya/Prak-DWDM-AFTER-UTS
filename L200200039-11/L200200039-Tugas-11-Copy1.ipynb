```python
from sklearn.cluster import KMeans

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
x = pd.read_csv("spotify_data.csv")
```

```python
x.head()
```

```python
x.drop(columns="Unnamed: 0",inplace=True)
x.drop(columns="acousticness",inplace=True)
x.drop(columns="artist",inplace=True)
x.isnull().sum()
```

```python
x = x.filter(["danceability", "energy"], axis = 1)
```

```python
wcss = []

for i in range (1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

#Melakukan plot untuk hasil sehingga bisa melakukan observasi terhadap elbow
plt.plot(range(1,15), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #WCSS = within cluster sum of squares
plt.show()
```

```python
sns.scatterplot(data = x, x="danceability", y="energy", c = ["green"])
```

```python
model = KMeans(n_clusters=5)
model.fit(x)
```

```python
print(model.cluster_centers_)
```

```python
sns.scatterplot (data = x, x="danceability", y="energy",  c=model.labels_, cmap = 'rainbow')
sns.scatterplot(x=model.cluster_centers_[:, 0], y=model.cluster_centers_[:, 1], c=['black'])
```
